```python
import pandas as pd
```

```python
#Print first 5 rows and last 5 rows
df = pd.read_csv('Automobile_data.csv')
print(df.head(5))
print(df.tail(5))
```

```python
#Find the highest price car amongst all the cars
df.nlargest(1,'price')[['company','price']]
```

```python
#Find the cars belonging to company 'toyota' 
df.loc[df['company']=='toyota']
```

```python
#Find the total number of cars per company and sort by descending order
df.groupby('company').count()['index'].sort_values(ascending =False)
```

```python
#Another way to compute the total number of values
df['company'].value_counts()
```

```python
#Find highest price car of each company
df.groupby('company').max('price')['price']
```

```python
#Find the average mileage of each company
df.groupby('company')['average_mileage'].mean()
```

```python
#Sort cars by the descending order of price 
df.sort_values('price', ascending = False).head(5)
```

```python
GermanCars = {'Company': ['Ford', 'Mercedes', 'BMV', 'Audi'], 'Price': [23845, 171995, 135925 , 71400]}
japaneseCars = {'Company': ['Toyota', 'Honda', 'Nissan', 'Mitsubishi '], 'Price': [29995, 23600, 61500 , 58900]}
Ge = pd.DataFrame(GermanCars)
Ja = pd.DataFrame(japaneseCars)
print(Ge)
print(Ja)
```

```python
#Concatenate two dataframes Ge and Ja 
pd.concat([Ge, Ja], keys = ["Germany","Japan"])
```

```python
Car_Price = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'Price': [23845, 17995, 135925 , 71400]}
car_Horsepower = {'Company': ['Toyota', 'Honda', 'BMV', 'Audi'], 'horsepower': [141, 80, 182 , 160]}
price = pd.DataFrame(Car_Price)
power = pd.DataFrame(car_Horsepower)
print(price)
print(power)
```

```python
#Merge the dataframe based on the common column Company
pd.merge(price,power, on = 'Company')
```

```python
#Pivot table to find the price for every category of body-style in every company
import numpy as np
pd.pivot_table(df, values ='price', index = ['company'], columns =['body_style'],aggfunc=np.mean)
```

```python
#Find the index number of the car with maximum value for average-mileage/length computation
((df['average_mileage'])/(df['length'])).idxmax()
```
