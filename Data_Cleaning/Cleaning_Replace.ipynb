```python
import pandas as pd
import numpy as np
```

```python
#Create a dataframe
df =pd.DataFrame({'a':['apple','orange','banana','grapes','cherry'],'b':['$20','$30','$60','$35','$65']})
df
```

DataFrame.replace(to_replace = ' ',value = ' ', inplace = ' ', method = ' ', regex = ' ') 


```python
#Replace one value | Value in the left and right side of the comma is considered to_replace = and value = respectively
df.replace('apple','guava')
```

```python
#List method 1 - Replace n number of values with n number of values | One to one mapping
df.replace(['apple', 'orange'],['guava','citrus'])
```

```python
#List method 2 - Replace muliple values with a single value | Many to one mapping
df.replace(['grapes','cherry'],'litchi')
```

```python
#Dictionary method 1  
#key:value like ({to_replace:value,to_replace:value})
df.replace({'apple':'guava'}) #Replace one element by passing dictionary 
df.replace({'apple':'guava','orange':'citrus'}) #Replace multiple elements by passing dictionary
```

```python
#Dictionary method 2 - Replace values in many columns 
#key:value like ({column_name:to_replace},{column_name:value})
df.replace({'a':'apple','b':'$60'}, 'Okay') 
df.replace({'a':'apple','b':'$60'},{'a':'melon','b':'60'})
```

```python
#Dictionary method 3 - Dictionary inside a dictionary | Replace values in a specific column
df.replace({'a':{'apple':'app','orange':'orang'}})
```

```python
#Replace the value with the suceeeding value | No need to pass value parameter
df.replace('banana',method='bfill')
```

```python
#regex = True need to mentioned to slice the characters and use regular expressions
df.replace('^a[a-z]*','Reg', regex = True) #start with 'a', a-z 0 or many times
df.replace('^or[a-z]+e$','Reg', regex = True) #start with 'or', a-z 1 or many times, end with e
df.replace('^b.[na]+','Reg', regex = True) #start with 'b', follow with a letter, letter a or n 1 or many times   
```

```python
#Regular expression using dict method 1
df.replace({'^c[a-z]*':'Reg','gr[a-z]*s$':'Reg'}, regex = True)
```

```python
#Regular expression using dict method 2 
df.replace({'a':'^a[pp]+.e$'},'appl', regex= True)
```

```python
#Regular expression using dict method 3
df.replace({'b':{'^\$2[0-9]*':'$90'}},regex=True)
```

```python
#Column cleanup and convert to float 
df1 = pd.read_csv('/content/Cleaning_Replace.csv')
df1
```

```python
#Remove '$' and ',' in MSRP | '\$' used since $ is used as an expression in Pandas 
df1.replace({'MSRP':{'\$':' ','\,':''}},regex = True,inplace=True) #Dict method 3
df1
```
