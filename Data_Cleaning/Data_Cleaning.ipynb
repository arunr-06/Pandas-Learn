```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Cleaning-Dataset.csv')
df
```

```python
# Display 'True' if the cell is null | isna is same as isnull
df.isnull()
```

```python
# Display 'False' if the cell is not null | notna is same as notnull
df.notnull()
```

```python
# Count the number of null values in each column
df.isnull().sum()
```

```python
# Replace the null values with 0
df.fillna(0)
```

```python
# Replace the null values with 0 in arrears
df['arrears'].fillna(0, inplace = True)
df
```

```python
# Replace the null values with the preceeding value
df.fillna(method='bfill')
```

```python
# Replace the null values with the succeeding value
df.fillna(method='ffill')
```

```python
# Replace the null value in SSLC with the mean value in SSLC 
df['SSLC'].fillna(df['SSLC'].mean(),inplace = True)
```

```python
df
```

```python
# fillna is used to replace the null values whereas dropna is used to remove the rows containing null values
# Remove the rows with null values
df.dropna()
```

```python
# Remove the rows which contains null values in all cells | By default how = 'any' and it will drop the rows containing null value in any one cell 
df.dropna(how='all')
```

```python
# Remove the rows which contains null values in either hsc or SSLC
df.dropna(subset=['hsc','SSLC'])
```

```python
# Remove the rows which contain null values in both hsc and SSLC
df.dropna(subset =['hsc','SSLC'], how = 'all')
```

```python
# Remove row 3
df.drop(3)
```

```python
# Remove column cmark
df.drop('cmark',axis=1)
```

```python
# Replace wrong values by desired values in college | Used to correct spelling mistakes/human error 
df['college'].replace({'IITM':'IIT Madras','SRM':'SRM University'}, inplace = True)
df
```
