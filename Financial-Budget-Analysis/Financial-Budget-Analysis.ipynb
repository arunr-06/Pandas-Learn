```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Read CSV file and show the first 5 rows

```python
df = pd.read_csv("India_budget_2021.csv")
df.head()
```

Find the row containing NaN values

```python
df.loc[df['Department /Ministry'].isnull()|df['Fund allotted(in ₹crores)'].isnull()]
```

Drop NaN values 

```python
df.dropna(inplace=True)
df
```

Find the index number of value 'GRAND TOTAL' in column Department /Ministry and store it in totalindex

```python
totalindex = df.loc[df['Department /Ministry']=='GRAND TOTAL'].index
totalindex
```

Drop the row with index number as in totalindex to remove 'GRAND TOTAL'

```python
df.drop(totalindex, inplace=True)
```

Select important departments and group the remaining department under Others
dfimp represents the important departments 

```python
dfimp = df.iloc[[0,8,11,14,18,23,41,42,43],:]
dfimp
```

dfother represents the other departments

```python
dfother = df.loc[~df['Department /Ministry'].isin(dfimp['Department /Ministry'])]
dfother
```

Find the sum of funds in dfother

```python
othersum = dfother['Fund allotted(in ₹crores)'].sum()
othersum
```

Create a dataframe with values - 'OTHERS' and the sum of funds in other departments 

```python
dfother1 = pd.DataFrame([{'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)':othersum}])
dfother1
```

Concatenate the dataframes

```python
dfcon = pd.concat([dfimp,dfother1],ignore_index=True)
dfcon
```

Display a bar chart

```python
dfcon.plot.bar(x='Department /Ministry', y= 'Fund allotted(in ₹crores)')
```

Display a pie chart

```python
fund = dfcon['Fund allotted(in ₹crores)']
dept = dfcon['Department /Ministry']
plt.pie(fund, labels = dept, autopct ='%1.1f%%', startangle=90, pctdistance=0.85)
central_circle = plt.Circle((0, 0), 0.5, color='white')
plt.gcf().gca().add_artist(central_circle)
plt.title("Distribution of Budget", fontsize=18)
```
