```python
import pandas as pd
import numpy as np
```

```python
#Import the csv file and read only 2 columns 
df = pd.read_csv('/content/Hotel-Occupancy.csv',usecols=['date','CommutativeOccupancy'])
df
```

```python
#Display the datatype and number of null values
df.info()
```

```python
#Replace the null values with the previous values using ffill 
#(If occupancy data of a day is not available, it is assumed to be unchanged from the previous day, hence ffill) 
df.fillna(method='ffill', inplace = True)
df
```

```python
#Find the difference betweent two successive columns and display it in a new column
#(Daily occupancy is found by subtracting the cummulative occupancy of consecutive days)
df['DailyOccupancy'] = df['CommutativeOccupancy'].diff()
df
```

```python
#Fill the na value of one column with the corresponding value of another column 
#(First day occupancy is 2)
df['DailyOccupancy'].fillna(df['CommutativeOccupancy'],inplace=True)
df
```
