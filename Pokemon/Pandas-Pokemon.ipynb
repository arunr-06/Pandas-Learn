Filter rows with string values 'Lit' or 'Fle' in 'Name'. Ignore cases using re.I.

```python
import re
df.loc[df['Name'].str.contains('Lit|Fle',flags = re.I)]
```

Filter rows with string values 'pi' at the beginning followed by any letters between a-z any number of times. Ignore cases using re.I.

```python
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I)]
```

Filter rows with value 'Grass' in 'Type 1' and post True in a new column 'Type 4'

```python
df.loc[df['Type 1'] == 'Grass', 'Type 4'] = True
df
```

Filter rows with multiple conditions 'HP' > 45 and 'Type 2'='Ghost' and post changes to multiple columns 'Attack' & 'Defense'

```python
df.loc[(df['HP']>45) & (df['Type 2']=='Ghost'),['Attack','Defense']]=[999,999]
df.loc[(df['Attack'])|(df['Defense']) == 999]
```

Filter rows with a specific string value  in 'Name' and post changes to fields in a new column'Type 3'

```python
df.loc[df['Name'].str.contains('^Pi[a-z]*'),'Type 3'] = 'Pi!'
```

Group by 'Type 1', sum the values - 'HP', 'Attack', 'Defense' for the corresponding 'Type 1' 

```python
df.groupby('Type 1').sum()[['HP','Attack','Defense']]
```

Group by two columns 'Type 1' and 'Type 2' and sort by descending order of the mean of 'HP' of each grouped set

```python
df.groupby(['Type 1','Type 2'])['HP'].mean().sort_values(ascending = False)
```

Group by 'Type 1' and find the sum, max and min of 'Sp. Atk'

```python
df.groupby('Type 2')['Sp. Atk'].agg([sum,max,min])
```

Group by 'Type 1' and display the mean of 'HP', max in 'Attack', and defense in 'Min' in the grouped dataset

```python
df.groupby('Type 1').agg({'HP':'mean','Attack':'max', 'Defense':'min'})
```

Replace value 'Grass' with 'Garden' in 'Type 1'

Replace the values 'Fire' into 'Fires' using map. Map is used to map all values of a Series based on a input.(NOTE: Map will mention NaN where it is not applicable. Replace would display the actual values even at places where it is not applicable.) 

```python
df['Type 1'].map({'Fire':'Fires'})
```

Transform the values of 'HP' by adding the max value  of 'HP' to 'HP' using tranform. Apply transform to a column/series and provide some inputs. Transformed column will have the same axis length as self.

```python
df['HP'].transform(lambda x:x+df['HP'].max())
```

Transform the values of 'HP' by displaying the square root and exponential.

```python
import numpy as np
df['HP'].transform([np.sqrt,np.exp])
```

Transform the values of 'HP' by posting it inside a string. na_action is used to avoid applying the function to null values. 

```python
df['HP'].transform('The HP of this pokemon is {}'.format, na_action = 'ignore')
```

Transform the values of 'Attack' after grouping it by 'Type 2' and summing the 'Attack' in each grouped dataset. It will display the the grouped values in every row.

```python
df.groupby('Type 2')['Attack'].transform(sum)
```

Transform the values after grouping it by 'Type 1' and adding 100 to its mean.

```python
df.groupby('Type 1').mean().transform(lambda x: x+100)['HP']
```

Pivot the rows and columns and find the aggregated mean of the value mentioned | Pivot converts data from long to wide 

```python
df.pivot_table(values='HP',index=['Type 1','Type 2'])
```

Pivot the rows and columns and find the aggregated sum of the values | If aggfunc is not mentioned, pandas will find the mean by default

```python
df.pivot_table(values='HP',index=['Type 1','Type 2'],aggfunc='sum')
```

Get the values in value_vars for each value in id_vars in individual rows | Melt converts data from wide to long

```python
df.melt(id_vars ='Type 1', value_vars =['Attack','Defense'])
```
