```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
#Creating a normal districution with paramters loc(mean), scale(standard deviation), size(population size) 
x = np.random.normal(loc=60,scale=3,size=100)
x
```

```python
#Finding the mean, minimum, maximum value & variance
x.mean(),x.min(),x.max(),x.var()
```

```python
#Reconfirming the variance using a python programme
addval = 0 
for i in x:
    val = (i-x.mean())**2
    addval = addval + val
    a = i-x.mean()
    b = a**2
    print(i,a,b,addval)
var = addval/len(x)
sd = math.sqrt(var)
print('Variance is', var)
print('Standard Deviation is', sd)
```

```python
#Creating a line plot
plt.plot(x)
```

```python
#Creating a histogram with bins referring the number of intervals
plt.hist(x)
```

```python
#Creating a kernel density estimation 
sns.kdeplot(x)
```

```python
#Creating a binomial distribution with n trials and p probability of occurence
x_bi = np.random.binomial(size=300,n=120,p=0.5)
x_bi
```

```python
plt.plot(x_bi)
```

```python
plt.hist(x_bi)
```

```python
sns.kdeplot(x_bi)
```

```python
#Creating a cross tab to display the value and the number of occurence
pd.set_option('display.max_columns',None)
pd.crosstab(index='count',columns=x_bi)
```
